```python
import pandas as pd
pd.set_option("display.max_colwidth", 200)

df = pd.DataFrame({'content': [
    "I love love love love this kitten",
    "I hate hate hate hate this keyboard",
    "I'm not sure how I feel about toast",
    "Sales of bad candy are up 200% in the third quarter",
    "Did you see the baseball game yesterday?",
    "The package was delivered late and the contents were broken",
    "Trashy television shows are some of my favorites",
    "I'm seeing a Kubrick film tomorrow, I hear not so great things about it.",
    "I find chirping birds irritating, but I know I'm not the only one",
]})
df
```

```python
scored = df.copy()

sentiment_pipeline = pipeline(model="cardiffnlp/twitter-xlm-roberta-base-sentiment")
results = sentiment_pipeline(df.content.tolist())
results = pd.DataFrame(results).add_prefix('twitter_multilanguage_')
scored = scored.join(results)

sentiment_pipeline = pipeline(model="LiYuan/amazon-review-sentiment-analysis")
results = sentiment_pipeline(df.content.tolist())
results = pd.DataFrame(results).add_prefix('amazon_')
scored = scored.join(results)

sentiment_pipeline = pipeline(model="mrm8488/distilroberta-finetuned-financial-news-sentiment-analysis")
results = sentiment_pipeline(df.content.tolist())
results = pd.DataFrame(results).add_prefix('finance_')
scored = scored.join(results)
```

```python
scored.style.background_gradient(cmap='RdYlGn', axis=None)
```

Now, **why do we care?** We probably don't care about sentiment analysis. We aren't a marketing research firm. **But this process has taught us four things:**

* Sentiment analysis is a classification problem (putting things into categories)
* Classification is done by models: some sort of computer thing that have a concept of the world inside of them
* Models are built by people feeding examples to them
* The model's output won't always be something you agree with, which is sometimes because they data it was trained on is the same as the data you're using it on
